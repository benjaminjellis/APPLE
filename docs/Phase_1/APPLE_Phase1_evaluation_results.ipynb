There is however, an important point to raise here: the best perfomring models are those with the most data. The date column in the table above is when these models were trained and saved. Over the course of phase 1, 3 models were built each week. Each model was then backtested on mined data and it's performance saved. The saved models were then deleted if they were underprofming on backtesting data. This means that the 7 retained models are from July.
